### Problem 2.3

$\textbf{(a)}$ Show
$$
\oint_{C_1} d\mathbf{s_1} \cdot \frac{\mathbf{r_1-r_2}}{|\mathbf{r_1-r_2}|^3} =0
$$

$\textbf{Solution:}$

From equation (2.21) in the text we have the identity,
$$
\nabla \frac{1}{|\mathbf{r_1-r_2}|} = - \frac{\mathbf{r_1-r_2}}{|\mathbf{r_1-r_2}|^3}
$$
So we can rewrite the integral as,
$$
- \oint_{C_1} d\mathbf{s_1} \cdot \nabla \frac{1}{|\mathbf{r_1-r_2}|}
$$
Using Stoke's theorem this becomes,
$$
- \oint_{S_1} d\mathbf{A} \cdot \nabla \times \nabla \frac{1}{|\mathbf{r_1-r_2}|}
$$
However, the curl of a gradient is identically zero. So we have,
$$
- \oint_{S_1} d\mathbf{A} \cdot \mathbf{0} = 0, 
$$
and the identity is proved.

$\textbf{(b)}$ The problem as written is a little confusing. It's not clear what the dot between the two integrals is referring to, and the second integral seems to involve a product of some kind between two vectors, the $d\mathbf{s_2}$ and the scaled vector difference of $\mathbf{r_1-r_2}$, which really makes no sense.

This is cleared up in the footnote on page 34 and in the wikipedia article on Ampere's force law:
https://en.wikipedia.org/wiki/Amp%C3%A8re%27s_force_law

The footnote makes it clear the dot product is between the $d\mathbf{s_1}$ and $d\mathbf{s_1}$. Rewriting it in the form used in the wikipedia article the force on wire 1 is,
$$
\mathbf{F_1} = \frac{\mu_0}{4\pi} \oint_{C_1}\oint_{C_2} (I_1 d\mathbf{s_1} \cdot I_2 d\mathbf{s_2}) \ \pmb{\mathscr{r}},
$$
where I've defined,
$$
\pmb{\mathscr{r}} =  \frac{\mathbf{r_1-r_2}}{|\mathbf{r_1-r_2}|^3},
$$
for convenience.

Reversing how the wikipedia article presents the material and using the triple cross product expansion lets us solve the problem.

Recall the BAC-CAB identity,
$$
A\times B\times C = B(A\cdot C) - C(A\cdot B).
$$
We can then write,
$$
d\mathbf{s_1} \times d\mathbf{s_2} \times \pmb{\mathscr{r}} = 
    d\mathbf{s_2} (d\mathbf{s_1}\cdot \pmb{\mathscr{r}}) - \bfscr (d\mathbf{s_1}\cdot d\mathbf{s_2})
$$
Isolating $\bfscr (d\mathbf{s_1}\cdot d\mathbf{s_2})$ and substituting into the equation for $\mathbf{F}_1$ we get,
$$
\mathbf{F_1} = -\frac{\mu_0}{4\pi} \oint_{C_1}\oint_{C_2} I_1 I_2 d\mathbf{s_2}\ (d\mathbf{s_1}\cdot \bfscr)
- d\mathbf{s_1} \times d\mathbf{s_2} \times \pmb{\mathscr{r}} \\
= \frac{\mu_0}{4\pi} \oint_{C_1}\oint_{C_2} I_1 I_2 d\mathbf{s_1} \times d\mathbf{s_2} \times \pmb{\mathscr{r}} -
\oint_{C_2} I_1 I_2 d\mathbf{s_2} 
\underbrace{\oint_{C_1} d\mathbf{s_1}\cdot \bfscr}_{= 0 \textrm{ by part (a)}} \\
= \oint_{C_1}I_1 d\mathbf{s_1} \times 
\underbrace{\frac{\mu_0}{4\pi} \oint_{C_2} I_2 d\mathbf{s_2} \times \pmb{\mathscr{r}}}_{= \mathbf{B_2(r)} \textrm{ by problem 2.2. See below.}}
$$

So,
$$\mathbf{F_1} = \oint_{C_1}I_1 d\mathbf{s_1} \times \mathbf{B_2(r)}$$
as required.

### Problem 2.4


$\textbf{Solution:}$

We have the magnetostatic version of Ampere's Law: 
$$\boxed{
\nabla\times \mathbf{B} = \mu_0 \mathbf{j}.
}$$

The conservation of charge for a general time-dependent charge density is:
$$\boxed{
\frac{\partial\rho}{\partial t} + \nabla\cdot\mathbf{j} = 0.}\ \ \ \ \ \ \ \ \ \ \ (*)
$$

Take the divergence of Ampere's Law,
$$
\underbrace{
\nabla\cdot\nabla\times\mathbf{B} }_{=0}
= \mu_0 \nabla\cdot\mathbf{j}.
$$
So we have $\nabla\cdot\mathbf{j} = 0$, which from the continuity equation implies,
$$
\frac{\partial\rho}{\partial t} = 0\ \Longrightarrow\ \textrm{always.}
$$
This is clearly incorrect.

So as suggested in the problem, add a term, $\mathbf{j}_D$, to Ampere's Law to correct for this. It's a try it and see approach where if it works, great. 

So now,
$$
\nabla\times \mathbf{B} = \mu_0 \mathbf{j} + \mathbf{j}_D,
$$
and we want to solve for $\mathbf{j}_D$.

Adding the term and taking the divergence again we get,
$$
\nabla\cdot\nabla\times\textbf{B} = \mu_0\nabla\cdot\textbf{j} + \nabla\cdot\textbf{j}_D.
$$
Using again the divergence of the curl equals zero we have,
$$
\nabla \cdot\textbf{j}_D = - \mu_0\nabla\cdot\textbf{j}.
$$
We can put the rhs in terms of change in charge density, $\rho$, by using the conservation of charge equation (*),
$$
\nabla \cdot\textbf{j}_D = \mu_0 \frac{\partial\rho}{\partial t}.\ \ \ \ \ \ \ \ \ \ \ (* *)
$$
We can then relate the change in charge density to the divergence of the electric field by Gauss' Law --- we know we want the electric field in there because we already know the answer, and we can get a divergence on the rhs to match the lhs's.
$$
\nabla\cdot\mathbf{E} = \frac{\rho}{\epsilon_0} \\
\Longrightarrow \nabla\cdot \frac{\partial\mathbf{E}}{\partial t} = \frac{1}{\epsilon_0} \frac{\partial\rho}{\partial t}.
$$
Substituting the $\frac{\partial\rho}{\partial t}$ into 
$(**)$, we have,
$$
\nabla \cdot\textbf{j}_D = \mu_0\epsilon_0 \nabla\cdot \frac{\partial\mathbf{E}}{\partial t}.
$$

Equating the arguments of the divergences we have,
$$
\textbf{j}_D = \mu_0\epsilon_0 \frac{\partial\mathbf{E}}{\partial t}.
$$

This gives the new general Ampere's Law,
$$
\boxed{
\nabla\times \mathbf{B} = \mu_0 \mathbf{j} + \mu_0\epsilon_0 \frac{\partial\mathbf{E}}{\partial t}.
}
$$

Note that $\mu_0\epsilon_0 = 1/c^2$, where $c$ is the speed of light. So the correction is very small and not measureable in a classical 1800's experimental setup.
